```python
import pandas as pd
```

```python
df=pd.read_csv('iris.csv')
```

```python
df.info()
```

seperate input output features

```python
X=df.drop(labels=['Species','Id'],axis=1)
Y=df[['Species']]
```

preprocessing


```python
#first try this method
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().set_output(transform='pandas')
X_pre=scaler.fit_transform(X)
```

```python
import pickle
with open('scaler.pkl','wb') as file1:   #wb means write binary
    pickle.dump(scaler,file1)
```

# check a file is added in notebook as scaler.pkl

```python
X_pre
```

# divide training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,Y,test_size=0.33,random_state=21)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytest.value_counts()
```

```python
#model buliding
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
#model evaluation
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
tr_pred=model.predict(xtrain)
ts_pred=model.predict(xtest)
```

```python
tr_pred[:5]
```

```python
ytrain.head()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cf=confusion_matrix(tr_pred,ytrain)
cf
```

```python
#above model is good model
```

```python
with open('model.pkl','wb') as file2:
    pickle.dump(model,file2)
```
